```python
from dataset import Dataset
from phasecorr import PhaseCorrelation
from plot import plot_images
```

# Cassini Saturn's F-ring

```python
ds = Dataset('saturn')
images = ds.sample_images()

names = [key for key in images]

pc = PhaseCorrelation(images[names[0]], images[names[1]], upscale_factor=1)
shifted_image = pc.transform()

plot_images(images[names[0]], images[names[1]], shifted_image)
```

# Sentinel-2 images

```python
ds = Dataset('sentinel2')
images = ds.sample_images()

names = [key for key in images]

pc = PhaseCorrelation(images[names[0]][:,:,:3], images[names[1]][:,:,:3], upscale_factor=5)
image = pc.transform()

plot_images(images[names[0]][:,:,:3], images[names[1]][:,:,:3], image[:, :, :3], cmap=None)
```
